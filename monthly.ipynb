```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objs as go
```

```python
df_english = pd.read_excel('dataset.xlsx')
df_hindu = pd.read_excel('dataset.xlsx')
```

```python
df_english = df_english.reset_index()
df_english
```

```python
# Convert the date column to a pandas datetime object
df_english['time'] = pd.to_datetime(df_english['time'])

# Set the date column as the DataFrame index
df_english.set_index('time', inplace=True)

# Resample the DataFrame to monthly frequency and calculate the average of other attributes
df_english_monthly = df_english.resample('M').mean()

# Reset the index to turn the date index back into a column
df_english_monthly = df_english_monthly.reset_index()

df_english_monthly
```

```python
# Monthly according to hindu calendar
df_hindu_monthly = df_hindu.groupby(['vikramSamvat', 'hinduMonthName'], sort=False).mean()
df_hindu_monthly = df_hindu_monthly.reset_index()
df_hindu_monthly
```

```python
# PLOT FOR HINDU CALENDAR

df = df_hindu_monthly
# Create traces for temperature and dewpoint
temp_trace = go.Scatter(x=df['vikramSamvat'].astype(str) + '-' + df['hinduMonthName'], y=df['temperature'], mode='lines', name='Temperature')
dew_trace = go.Scatter(x=df['vikramSamvat'].astype(str) + '-' + df['hinduMonthName'], y=df['dewpoint'], mode='lines', name='Dewpoint')
relative_humidity_percentage_trace = go.Scatter(x=df['vikramSamvat'].astype(str) + '-' + df['hinduMonthName'], y=df['relativeHumidityPercentage'], mode='lines', name='Relative Humidity Percentage')
windspeed_trace = go.Scatter(x=df['vikramSamvat'].astype(str) + '-' + df['hinduMonthName'], y=df['windspeed'], mode='lines', name='Windspeed')
soil_moisture_trace = go.Scatter(x=df['vikramSamvat'].astype(str) + '-' + df['hinduMonthName'], y=df['soil_moisture'], mode='lines', name='Soil Moisture')

# Create a layout with x and y axis labels
layout = go.Layout(title='Line Graph for Hindu Calendar', xaxis_title='Hindu Year and Month', yaxis_title='Attributes', xaxis=dict(tickangle=45))

# Create a figure with both traces and the layout
fig = go.Figure(data=[temp_trace, dew_trace, relative_humidity_percentage_trace, windspeed_trace, soil_moisture_trace], layout=layout)

# Show the figure
fig.show()
```

```python
# PLOT FOR English CALENDAR

df = df_english_monthly
# Create traces for temperature and dewpoint
temp_trace = go.Scatter(x=df['time'].astype(str), y=df['temperature'], mode='lines', name='Temperature')
dew_trace = go.Scatter(x=df['time'].astype(str), y=df['dewpoint'], mode='lines', name='Dewpoint')
relative_humidity_percentage_trace = go.Scatter(x=df['time'].astype(str), y=df['relativeHumidityPercentage'], mode='lines', name='Relative Humidity Percentage')
windspeed_trace = go.Scatter(x=df['time'].astype(str), y=df['windspeed'], mode='lines', name='Windspeed')
soil_moisture_trace = go.Scatter(x=df['time'].astype(str), y=df['soil_moisture'], mode='lines', name='Soil Moisture')

# Create a layout with x and y axis labels
layout = go.Layout(title='Line Graph for English Calendar', xaxis_title='Hindu Year and Month', yaxis_title='Attributes', xaxis=dict(tickangle=45))

# Create a figure with both traces and the layout
fig = go.Figure(data=[temp_trace, dew_trace, relative_humidity_percentage_trace, windspeed_trace, soil_moisture_trace], layout=layout)

# Show the figure
fig.show()
```

```python
hdf = df_hindu_monthly
edf = df_english_monthly

# Create traces for Hindu Calendar
temp_box_h = go.Box(y=hdf['temperature'], name='Temperature')
dew_box_h = go.Box(y=hdf['dewpoint'], name='Dewpoint')
relative_humidity_percentage_h = go.Box(y=hdf['relativeHumidityPercentage'], name='Relative Humidity Percentage')
windspeed_h = go.Box(y=hdf['windspeed'], name='Windspeed')
soil_moisture_h = go.Box(y=hdf['soil_moisture'], name='Soil Moisture')

# Create traces for English Calendar
temp_box_e = go.Box(y=edf['temperature'], name='Temperature')
dew_box_e = go.Box(y=edf['dewpoint'], name='Dewpoint')
relative_humidity_percentage_e = go.Box(y=edf['relativeHumidityPercentage'], name='Relative Humidity Percentage')
windspeed_e = go.Box(y=edf['windspeed'], name='Windspeed')
soil_moisture_e = go.Box(y=edf['soil_moisture'], name='Soil Moisture')

# Create a layout with x and y axis labels
layout = go.Layout(title='Box Plots', yaxis_title='Attributes', xaxis=dict(tickangle=45))

# Create a figure with both box plots and the layout
hfig = go.Figure(data=[temp_box_h, dew_box_h, relative_humidity_percentage_h, windspeed_h, soil_moisture_h], layout=layout)
efig = go.Figure(data=[temp_box_e, dew_box_e, relative_humidity_percentage_e, windspeed_e, soil_moisture_e], layout=layout)

# Show the figure
hfig.show()
efig.show()
```

```python
# English Months Line graph

df = df_english.reset_index()

# Convert the date column to a datetime type
df['time'] = pd.to_datetime(df['time'], format='%d-%m-%Y')

# Group the data by month
grouped = df.groupby(df['time'].dt.strftime('%B'))

# Create a separate figure for each month
for month_name, month_data in grouped:
    # Group the data by year
    year_grouped = month_data.groupby(month_data['time'].dt.year)

    # Create a list to store the traces
    traces = []

    # Create a trace for each year
    for year, year_data in year_grouped:
        temp_trace = go.Scatter(x=year_data['time'], y=year_data['temperature'],
                                mode='lines', name=str(year))
        traces.append(temp_trace)

    # Create a layout with x and y axis labels and a title
    layout = go.Layout(title=f'Temperature in {month_name}', xaxis_title='time', yaxis_title='Temperature')

    # Create a figure with the temperature traces and the layout
    fig = go.Figure(data=traces, layout=layout)

    # Set the mode of the temperature trace to lines
    fig.update_traces(mode='lines')

    # Show the figure
    fig.show()
```
